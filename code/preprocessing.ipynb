```python
import SimpleITK as sitk
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
path = '/nas/dataset/minhyeok/LDCT/C002/12-23-2021-NA-NA-62464'

dif = list()

full = os.listdir(path)[0]
low = os.listdir(path)[1]

for i in range(280):
    fimg = os.listdir(f'{path}/{full}')[i]
    fdcm = sitk.ReadImage(f'{path}/{full}/{fimg}')
    farr = sitk.GetArrayFromImage(fdcm)
    farr = (farr - np.min(farr))/ (np.max(farr)- np.min(farr))
    limg = os.listdir(f'{path}/{low}')[i]
    ldcm = sitk.ReadImage(f'{path}/{low}/{limg}')
    larr = sitk.GetArrayFromImage(ldcm)
    larr = (larr - np.min(larr)) / (np.max(larr)-np.min(larr))
    diff = farr - larr
    dif.append(diff)
    
    
# cn=0
# for a in os.listdir(f'{path}/{os.listdir(path)[1]}'):
#     # print(a)
#     cn+=1
# print(cn)
    
```
